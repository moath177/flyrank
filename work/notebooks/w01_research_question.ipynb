## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd, numpy as np
import os, sys, subprocess
IN_COLAB = "google.colab" in sys.modules
REPO_URL = "https://github.com/flyrank-bih/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"
if IN_COLAB:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
    os.chdir(REPO_DIR)
    subprocess.run([sys.executable, "-m", "pip", "install", "-q", "-r", "requirements.txt"], check=True)
elif os.path.basename(os.getcwd()) == "notebooks":
    os.chdir("..")  # move from notebooks/ to the repo root
print("Working dir:", os.getcwd())
assert os.path.exists("data/raw/content_refresh_anonymized.csv"), "starter CSV not found — are you at the repo root?"
print("Starter data found. You're ready.")
print("="*40)


df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
# 1. Pages that already look like CTR opportunities (visible, decently positioned, weak CTR)
low_ctr = df[(df["impressions_90d"] >= 500) & (df["avg_position"] > 0) &
             (df["avg_position"] <= 20) & (df["ctr"] < 0.5)]
print(f"Low-CTR visible pages: {len(low_ctr):,} / {len(df):,} ({len(low_ctr)/len(df):.1%})")

# 2. Pages with enough sessions but weak engagement
low_eng = df[(df["sessions_90d"] >= 30) &
             ((df["engagement_rate"] < 30) | (df["scroll_rate"] < 30))]
print(f"Low-engagement visible pages: {len(low_eng):,} / {len(df):,} ({len(low_eng)/len(df):.1%})")

# 3. Expected CTR depends heavily on position tier -- raw CTR can't be compared across tiers
valid_pos = df[df["avg_position"] > 0]
print(valid_pos.groupby("position_tier")["ctr"].mean().sort_values(ascending=False))
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

#What I can claim:
page-level CTR compared against the average CTR of other pages in the same position tier is an observed, position-adjusted gap, measured directly from the starter dataset. Ranking pages by this gap produces a reviewable, decision-support priority list: a large, persistent gap combined with enough impression volume is a reasonable, directional signal to flag a page for human review — not a certainty.

#What I can not claim:
that a low CTR is caused specifically by a weak title or meta description — the gap could equally come from competing SERP features, a mismatch between search intent and content, seasonality, or noise on low-volume pages. I also cannot claim that editing a page will cause its CTR to improve; proving that would require an actual experiment, which this data alone does not provide. Finally, this analysis says nothing about Google's ranking algorithm itself — only about observed clicks relative to observed position.
